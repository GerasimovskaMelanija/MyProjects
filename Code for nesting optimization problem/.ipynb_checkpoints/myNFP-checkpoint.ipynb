```python
import numpy as np
from shapely.geometry import Point, Polygon
import random
import matplotlib.pyplot as plt
import matplotlib.patches

c1 = [[0, 86], [966, 142], [1983, 0], [2185, 238], [2734, 217],[3000, 767],[2819, 900], [2819, 1360], [3000, 1439], [2734, 2043], [2185, 2022], [1983, 2260], [966, 2118], [0, 2174]]
#c1=[[0,0],[5,0],[0,5]]
p1=Polygon(c1)

c3 = [[0, 0], [3034, 0], [3034, 261], [0, 261]]
#c3=[[5,3],[8,3],[8,0],[5,0]]
p3=Polygon(c3)
while p1.intersects(p3):
    for k in range(len(c1)):
        c1[k][0] = c1[k][0] + 1
        c1[k][1] = c1[k][1] + 1
    p1 = Polygon(c1)
print(c1)
def NFP(poly1,poly2):
    NFP_koord =[]
    c11=np.array(poly1)
    c33=np.array(poly2)
    for i in range(len(c11)):
        for j in range(len(c33)):
            NFP_koord.append(c11[i]+c33[j])

    for i in range(len(NFP_koord)):
        NFP_koord[i]  = NFP_koord[i].tolist()
    #print(NFP_koord)
    
    return NFP_koord
nfp = NFP(c1,c3)

#print(len(nfp))
plt.figure()
for j in range (len(nfp)):
    
    plt.scatter(nfp[j][0],nfp[j][1])
p11 = c1
p22 = c3
p11.append(p11[0])
p22.append(p22[0])
x111, y111 = zip(*p11)
x222, y222 = zip(*p22)
nfp.append(nfp[0])
n1,h1 = zip(*nfp)
plt.figure(figsize = (8.0, 8.0))
plt.plot(x111,y111, 'b' ,label ='Prv poligon')
plt.plot(x222,y222, 'g',label ='Vtor poligon')
plt.figure(figsize = (8.0, 8.0))

plt.plot(n1,h1,'r',label='Minkowski sum')

```
