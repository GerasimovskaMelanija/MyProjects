```python
import numpy as np
import pandas as pd
from shapely.geometry import Point, Polygon
import random
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('fu.xlsx')
data.head(6)
size = data.iloc[1]['Unnamed: 2']
width = data.iloc[2]['Unnamed: 4']
koordinati = [] 
for i in range(len(data.index)):
    if i >=4 :
        if pd.notnull(data.iloc[i]['Unnamed: 0']) == True:
            t = []
            x = []
            y = []
            for j in range(1,len(data.columns)):
                if pd.notnull(data.iloc[i]['Unnamed: ' + str(j)]) == True:
                    
                    x.append(data.iloc[i]['Unnamed: '+ str(j)])
                    y.append(data.iloc[i+1]['Unnamed: ' + str(j)])
            for m in range(len(x)):
                if m>0:
                    
                    t.append([x[m],y[m]])
            koordinati.append(t)
Poligoni = []
for i in range(len(koordinati)):
    Poligoni.append(Polygon(koordinati[i]))
```
