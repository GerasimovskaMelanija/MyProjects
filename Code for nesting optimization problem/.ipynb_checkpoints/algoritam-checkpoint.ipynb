#### Функција за пресметување на NFP:

```python
def NFP(poly1, poly2): #Како аргументи ги прима координатите на двата полигони
    p1 = sg.Polygon(poly1)
    p2 = sg.Polygon(poly2)
    result = minkowski.minkowski_sum(p1, p2)
    return result
    
```
